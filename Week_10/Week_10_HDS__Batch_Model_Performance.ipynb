The line graph shows the change of accuracy, precision and recall during the five stages of model development and evaluation:

1.   Dummy

* Dummy Step:

* KNN-2 (Initial KNN Prediction):

* KNN-3-1 (Retraining):
Precision ~0.43 and accuracy ~0.73 suffer from an unexpected drop, while recall peaks ~1.0. This may indicate over fitting in a particular class or class imbalance, where the model becomes extremely sensitive to positives, but is less precise since many positives exist.

* KNN-3-2 (Retrained Prediction):
It stabilizes and improves again. Accuracy recovers ~0.95 and precision ~0.90 and recall ~0.91 recover to strong levels. This phase shows that retraining did indeed alleviate previous instability and that model performance was balanced on precision recall trade offs.
